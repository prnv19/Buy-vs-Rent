## Select Date Range: 2018-08-31 to 2024-08-31

```python
def filter_dates(df, cols = 5):
    date_columns = df.columns[cols:]  # Date columns
    date_columns = pd.to_datetime(date_columns, errors='coerce')
    
    start_date = datetime(2018, 8, 31)
    end_date = datetime(2024, 8, 31)
    
    # Create a mask for the desired date range
    date_mask = (date_columns >= start_date) & (date_columns <= end_date)
    
    # Get the names of the columns that match the date mask
    filtered_date_columns = date_columns[date_mask].strftime('%Y-%m-%d')

    # Combine the first 5 columns with the filtered date columns
    filtered_df = df.loc[:, [*ori_df.columns[:cols], *filtered_date_columns]]
    
    return filtered_df
```

```python
ori_df = filter_dates(ori_df)
ori_df
```

```python
hvi_df = filter_dates(hvi_df)
hvi_df
```

```python
msp_df = filter_dates(msp_df)
msp_df
```

```python
mlp_df = filter_dates(mlp_df)
mlp_df
```

## Fill missing values

```python
#interpolate rows

ori_df.interpolate(method='linear', axis=0, inplace=True)
hvi_df.interpolate(method='linear', axis=0, inplace=True)
msp_df.interpolate(method='linear', axis=0, inplace=True)
mlp_df.interpolate(method='linear', axis=0, inplace=True)
```

## Combined dataset

```python
#Save the cleaned datasets
ori_df.to_csv('Datasets/cleaned/observed_rent_index.csv', index=False)
hvi_df.to_csv('Datasets/cleaned/home_value_index.csv', index=False)
msp_df.to_csv('Datasets/cleaned/median_sale_price.csv', index=False)
mlp_df.to_csv('Datasets/cleaned/median_list_price.csv', index=False)
```

```python
ori_df = pd.read_csv('Datasets/cleaned/observed_rent_index.csv')

ori_df['observed_rent_index'] = pd.NA
for i in range(len(ori_df)):
    #Convert the row to a list and add the values to the new column
    ori_df['observed_rent_index'][i] = [float(x) for x in list(ori_df.iloc[i, 5:len(ori_df.columns)-1])]

#Drop the date columns
ori_df.drop(ori_df.columns[5:len(ori_df.columns)-1], axis=1, inplace=True)
ori_df
```

```python
hvi_df = pd.read_csv('Datasets/cleaned/home_value_index.csv')

hvi_df['home_value_index'] = pd.NA
for i in range(len(hvi_df)):
    #Convert the row to a list and add the values to the new column
    hvi_df['home_value_index'][i] = [float(x) for x in list(hvi_df.iloc[i, 5:len(hvi_df.columns)-1])]

#Drop the date columns
hvi_df.drop(hvi_df.columns[5:len(hvi_df.columns)-1], axis=1, inplace=True)
hvi_df
```

```python
msp_df = pd.read_csv('Datasets/cleaned/median_sale_price.csv')

msp_df['median_sale_price'] = pd.NA
for i in range(len(hvi_df)):
    #Convert the row to a list and add the values to the new column
    try:
        msp_df['median_sale_price'][i] = [float(x) for x in list(msp_df.iloc[i, 5:len(msp_df.columns)-1])]
    except:
        pass

#Drop the date columns
msp_df.drop(msp_df.columns[5:len(msp_df.columns)-1], axis=1, inplace=True)
msp_df
```

```python
mlp_df = pd.read_csv('Datasets/cleaned/median_list_price.csv')

mlp_df['median_list_price'] = pd.NA
for i in range(len(hvi_df)):
    #Convert the row to a list and add the values to the new column
    mlp_df['median_list_price'][i] = [float(x) for x in list(mlp_df.iloc[i, 5:len(mlp_df.columns)-1])]

#Drop the date columns
mlp_df.drop(mlp_df.columns[5:len(mlp_df.columns)-1], axis=1, inplace=True)
mlp_df
```

```python
#Find the intersection of the 4 datasets using "RegionName" column
dfs = [ori_df, hvi_df, msp_df, mlp_df]
df_final = pd.concat([df.set_index('RegionName') for df in dfs], axis=1, join='inner').reset_index()

columns_to_drop = [ 'RegionID', 'SizeRank', 'RegionType', 'RegionID', 'SizeRank', 'RegionType','RegionID', 'SizeRank', 'RegionType','RegionID', 'SizeRank', 'RegionType']
#Drop the columns
df_final.drop(columns_to_drop, axis=1, inplace=True)

#Merge duplicate columns
df_final = df_final.loc[:,~df_final.columns.duplicated()]

df_final
```

## Region Selection

```python
# Select rows with column as CO
df_final = df_final[df_final['StateName'] == 'CO']

#Remove ', CO' from the 'RegionName' column
df_final['RegionName'] = df_final['RegionName'].str.replace(', CO', '')
df_final['RegionName'].iloc[10] = 'Canon City'

df_final
```
