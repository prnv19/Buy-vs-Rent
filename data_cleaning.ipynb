```python
# ori_df = pd.read_csv('Datasets/cleaned/observed_rent_index.csv')

ori_df['observed_rent_index'] = pd.NA
for i in range(len(ori_df)):
    #Convert the row to a list and add the values to the new column
    ori_df['observed_rent_index'][i] = [float(x) for x in list(ori_df.iloc[i, 5:len(ori_df.columns)-1])]

#Drop the date columns
ori_df.drop(ori_df.columns[5:len(ori_df.columns)-1], axis=1, inplace=True)
ori_df
```

```python
# hvi_df = pd.read_csv('Datasets/cleaned/home_value_index.csv')

hvi_df['home_value_index'] = pd.NA
for i in range(len(hvi_df)):
    #Convert the row to a list and add the values to the new column
    hvi_df['home_value_index'][i] = [float(x) for x in list(hvi_df.iloc[i, 5:len(hvi_df.columns)-1])]

#Drop the date columns
hvi_df.drop(hvi_df.columns[5:len(hvi_df.columns)-1], axis=1, inplace=True)
hvi_df
```

## Region Selection

```python
# Select rows with column as CO
ori_df = ori_df[ori_df['StateName'] == 'CO']

#Remove ', CO' from the 'RegionName' column
ori_df['RegionName'] = ori_df['RegionName'].str.replace(', CO', '')
ori_df['RegionName'].iloc[10] = 'Canon City'

#Remove columns
ori_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1, inplace=True)

#Reset the index
ori_df.reset_index(drop=True, inplace=True)

ori_df
```

```python
# Select rows with column as CO
hvi_df = hvi_df[hvi_df['StateName'] == 'CO']

#Remove ', CO' from the 'RegionName' column
hvi_df['RegionName'] = hvi_df['RegionName'].str.replace(', CO', '')
hvi_df['RegionName'].iloc[10] = 'Canon City'

#Remove columns
hvi_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1, inplace=True)

#Reset the index
hvi_df.reset_index(drop=True, inplace=True)

hvi_df
```

```python
#Save the final dataset
ori_df.to_pickle('Datasets/observed_rent_index_co.pkl')
hvi_df.to_pickle('Datasets/home_value_index_co.pkl')
```
